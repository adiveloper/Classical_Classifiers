```python
def reshaper(X):
    x=[]
    for i in range(len(X)):
        v=[]
        a=X[i].reshape(X.shape[3],-1)
        for c in range(len(a)):
            v.append(sum(a[c])/(X.shape[1]*X.shape[2]))
        x.append(np.array(v).T)
    return np.array(x),np.array(v)
```
