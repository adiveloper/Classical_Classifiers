```python
datadir = 'micrograph'
columns = ['micrograph_id','path','micron_bar','micron_bar_units','micron_bar_px',
           'magnification','detector','sample_key','contributor_key','primary_microconstituent']
data = pd.read_csv('micrograph.csv')
#pd.set_option('display.max_colwidth', -1)
data.head()
#X    =   train[0:,1:2]
#Y =      train[0:,9:10]
#Y.shape
```

```python
print(data.iloc[50])
def datagen(data_dir, df):
    image_path = []
    label = []
    for i in range(len(data)):
        indexed_data = data.iloc[i]
        if(indexed_data[9]=='pearlite' or indexed_data[9]=='network' or indexed_data[9]=='spheroidite'):
            X = indexed_data[1]
            image_path.append(os.path.join(data_dir, X.strip()))
            label.append(indexed_data[9])
        image_paths = np.asarray(image_path)
        labels = np.asarray(label)
    return image_paths,labels

image_paths, labels = datagen(datadir, data)
```

```python
le = preprocessing.LabelEncoder()
le.fit(['pearlite','network','spheroidite'])
labels = le.transform(labels)
labels = to_categorical(labels)
```

```python
print(labels)
```

```python
def pimg(paths,labels):
    img = mpimg.imread(paths)
    if(img.shape !=  (522, 645)):
        img = np.resize(img[200:424,200:424,0],(224,224))
        img = np.array(cv2.cvtColor(img,cv2.COLOR_GRAY2RGB))
    else:
        img = np.resize(img[200:424,200:424],(224,224))
        img = np.array(cv2.cvtColor(img,cv2.COLOR_GRAY2RGB))
    
    return img,labels
```

```python
X_train, x_valid,y_train, y_valid = train_test_split(image_paths,labels, test_size = 0.1)
print('Training Samples {} \nValidation Samples {}'.format(len(X_train), (len(x_valid))))
```

```python
def batch_generator(image_paths, labels,batch_size, istraining):
    while True:
        batch_img = []
        batch_labels = []
        
        for i in range (batch_size):
            
            
            if istraining:
                random_index = random.randint(0,len(image_paths)-100)
                im,lab = pimg(image_paths[random_index], labels[random_index])

                
            else:
                random_index = random.randint(len(image_paths)-100,len(image_paths)-1)
                im,lab = pimg(image_paths[random_index], labels[random_index])         
            batch_img.append(im)
            batch_labels.append(lab)
        yield (np.asarray(batch_img), np.asarray(batch_labels))

```

```python
def CNNmodel():
    model = Sequential()
    model.add(Conv2D(24, (5, 5), subsample = (2,2), input_shape = (200,200,1), activation = 'relu'))
    model.add(Conv2D(36, (5, 5), subsample = (2,2), activation = 'relu'))
    model.add(Conv2D(48, (5, 5), subsample = (2,2), activation = 'relu'))
    model.add(Conv2D(64, (3, 3), activation = 'relu'))
    model.add(Conv2D(64, (3, 3), activation = 'relu'))

    
    model.add(Flatten())
    model.add(Dense(100, activation = 'relu'))

    
    model.add(Dense(50, activation = 'relu'))

    model.add(Dense(10, activation = 'relu'))
  
    model.add(Dense(3,activation = 'softmax'))
    adam = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=True)
    model.compile(loss = 'mse', optimizer = adam)
    return model
```

```python
model = CNNmodel() 
model.compile(Adam(lr=0.0001), loss = 'categorical_crossentropy',metrics=['accuracy'])
```

```python
h = model.fit_generator(batch_generator(X_train, y_train,32, 1), 
                        steps_per_epoch = 30, epochs = 50,
                        validation_data = batch_generator(x_valid,y_valid, 32 , 0), 
                        validation_steps = 20, verbose = 1, shuffle = 1)
```

```python
plt.plot(h.history['loss'])
plt.plot(h.history['val_loss'])
plt.legend(['training loss', 'val_loss'])
plt.title('loss')
```
